```python
import pandas as pd
```

```python
df1 = pd.read_excel("/content/example_ventas.xlsx")
```

```python
df1
```

```python
# Agrupar por 'Producto' y sumar la columna 'Ventas'
ventas_por_producto = df1.groupby('Producto')['Ventas'].sum().reset_index()

# Mostrar el resultado
print(ventas_por_producto)
```

```python
# Agrupar por 'Region' y sumar la columna 'Ventas'
ventas_por_region = df1.groupby('Region')['Ventas'].sum().reset_index()

# Mostrar el resultado
print(ventas_por_region)
```

```python
# Definir la tasa de conversión de euros a pesos colombianos
tasa_conversion = 4600  # Ejemplo: 1 euro = 4600 COP

# Crear una copia del DataFrame df1
df_convertido = df1.copy()

# Convertir el valor de 'Ventas' de euros a pesos colombianos en una nueva columna
df_convertido['Valor_PesoCol'] = df_convertido['Ventas'] * tasa_conversion

# Mostrar el DataFrame actualizado
print(df_convertido.head())
```

```python
# Filtrar el DataFrame para que solo muestre las filas donde el vendedor sea 'David'
df_david = df1[df1['Vendedor'] == 'David']

# Mostrar el DataFrame filtrado
print(df_david)
```

```python
# Filtrar las filas donde las ventas son mayores a 10,000
df_ventas_mayores_10000 = df1[df1['Ventas'] > 10000]

# Ordenar el DataFrame filtrado por la columna 'Ventas' de mayor a menor
df_ventas_mayores_10000 = df_ventas_mayores_10000.sort_values(by='Ventas', ascending=False)

# Mostrar el DataFrame filtrado y ordenado
print(df_ventas_mayores_10000)
```

```python
df2 = df1.drop('Vendedor', axis=1)
print(df2)
```
